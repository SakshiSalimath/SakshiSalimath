```python
#find all racing games critic score >=90
racing_games = video_games_data.loc[(video_games_data.Genre=='Racing')
                                          &
                                    (video_games_data.Critic_Score>=90)]
```

```python
#find all action and racing games  and both should have critic score>=90
racing_action_games = video_games_data.loc[(video_games_data.Genre=='Racing')
                                                |
                                           (video_games_data.Genre=='Action')
                                           &
                                           (video_games_data.Critic_Score>=90)]
```
